```python
import cv2 as cv
from IPython.display import display, Image
```

## Sample image

```python
img = cv.imread("../sample/cat.jpg")
display(Image("../sample/cat.jpg", width=200))
```

## Grayscale by splitting color channels

```python
b_channel, g_channel, r_channel = img[:,:,0], img[:,:,1], img[:,:,2]

cv.imwrite("../tmp/cat_b_channel.jpg", b_channel)
cv.imwrite("../tmp/cat_g_channel.jpg", g_channel)
cv.imwrite("../tmp/cat_r_channel.jpg", r_channel)

display(
    Image("../tmp/cat_b_channel.jpg", width=200),
    Image("../tmp/cat_g_channel.jpg", width=200),
    Image("../tmp/cat_r_channel.jpg", width=200),
)
```

## Grayscale by using algorithm


$$ \text{Formula 1: } Y = (R + G + B) / 3 $$

```python
white_f1 = (r_channel + g_channel + b_channel) / 3
cv.imwrite("../tmp/cat_white_f1.jpg", white_f1)
display(Image("../tmp/cat_white_f1.jpg", width=200))
```

The simplest way to grayscale image is balancing all the 3 colors. However it leads to darker image which is not the gray we want 

$$ \text{Formula 2: } Y = 0.299 * R + 0.587 * G + 0.114 * B $$

```python
white_f2 = (0.299 * r_channel) + (0.587 * g_channel) + (0.114 * b_channel)
cv.imwrite("../tmp/cat_white_f2.jpg", white_f2)
display(Image("../tmp/cat_white_f2.jpg", width=200))
```

By putting different weights on each channel, we can create a more natural grayscale image.

## Convert back to color image

```python
new_img = cv.merge([b_channel, g_channel, r_channel])
cv.imwrite("../tmp/cat_merge_channel.jpg", new_img)
display(Image("../tmp/cat_merge_channel.jpg", width=200))
```

## Conclusion

We can easily convert back to color image from grayscale if we have all 3 grayscale images of the 3 color channels, else it's very difficult because a bunch of data was missing on the merging grayscale images process.
